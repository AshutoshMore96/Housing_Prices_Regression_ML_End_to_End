```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
train_df = pd.read_csv("/Users/ashutosh/PycharmProjects/Housing_Prices_Regression_ML_End_to_End/data/processed/feature_engineered_train.csv")
eval_df  = pd.read_csv("/Users/ashutosh/PycharmProjects/Housing_Prices_Regression_ML_End_to_End/data/processed/feature_engineered_eval.csv")
```

#### Declare the target feature and other features

```python
target = "price"
X_train = train_df.drop(columns=[target, "lat", "lng"])
y_train = train_df[target]

X_eval = eval_df.drop(columns=[target, "lat", "lng"])
y_eval = eval_df[target]
```

```python
X_train
```

```python
print("Any NaN in X_train?", X_train.isna().any().any())
print("Any NaN in X_eval?", X_eval.isna().any().any())
print("*" * 75)

print("Any inf in X_train?", np.isinf(X_train).any().any())
print("Any inf in X_eval?", np.isinf(X_eval).any().any())
print("*" * 75)

print("NaNs per column in X_train:")
print(X_train.isna().sum().sort_values(ascending=False).head(20))
print("*" * 75)

print("Infs per column in X_train:")
print(np.isinf(X_train).sum().sort_values(ascending=False).head(20))
```

#### Standardize the features from training set by 'fit' method and 'transform' method on evaluation set

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_eval_scaled  = scaler.transform(X_eval)
```

#### Train the models followed by evaluation

###### 1. Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
y_pred_lr = lr.predict(X_eval_scaled)

print("Linear Regression:")
print("MeanAbsoluteError:", mean_absolute_error(y_eval, y_pred_lr))
print("RootMeanSquaredError:", np.sqrt(mean_squared_error(y_eval, y_pred_lr)))
print("R²:", r2_score(y_eval, y_pred_lr))
```

###### 2. Ridge Regression

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, y_train)
y_pred_ridge = ridge.predict(X_eval_scaled)

print("\nRidge Regression:")
print("MeanAbsoluteError:", mean_absolute_error(y_eval, y_pred_ridge))
print("RootMeanSquaredError:", np.sqrt(mean_squared_error(y_eval, y_pred_ridge)))
print("R²:", r2_score(y_eval, y_pred_ridge))
```

###### 3. Lasso Regression

```python
lasso = Lasso(alpha=0.1)
lasso.fit(X_train_scaled, y_train)
y_pred_lasso = lasso.predict(X_eval_scaled)

print("\nLasso Regression:")
print("MeanAbsoluteError:", mean_absolute_error(y_eval, y_pred_lasso))
print("RootMeanSquaredError:", np.sqrt(mean_squared_error(y_eval, y_pred_lasso)))
print("R²:", r2_score(y_eval, y_pred_lasso))
```

###### 4. ElasticNet

```python
elastic = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic.fit(X_train_scaled, y_train)
y_pred_elastic = elastic.predict(X_eval_scaled)

print("\nElasticNet Regression:")
print("MeanAbsoluteError:", mean_absolute_error(y_eval, y_pred_elastic))
print("RootMeanSquaredError:", np.sqrt(mean_squared_error(y_eval, y_pred_elastic)))
print("R²:", r2_score(y_eval, y_pred_elastic))
```
